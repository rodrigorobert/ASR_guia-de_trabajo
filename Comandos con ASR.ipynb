```python
#Funcion para 'repetir'

def talk(text):
    engine.say(text)
    engine.runAndWait()

listener = sr.Recognizer()

try:
  with sr.Microphone() as source:
    print('Escuchando...')
    voice = listener.listen(source)
    rec = listener.recognize_google(voice, language='es-AR')
    talk(rec)
except:
    pass
```

```python
#Función para reconocer un nombre o palabra clave

name = 'juanita'

def talk(text):
    engine.say(text)
    engine.runAndWait()

listener = sr.Recognizer()

try:
  with sr.Microphone() as source:
    print('Escuchando...')
    voice = listener.listen(source)
    rec = listener.recognize_google(voice, language='es-AR')
    rec = rec.lower()
    if name in rec:
        talk(rec)
except:
    pass
```

```python
#Función para ejecutar un comando

def talk(text):
    engine.say(text)
    engine.runAndWait()

listener = sr.Recognizer()

def listen():
    try:
        with sr.Microphone() as source:
            print('Escuchando...')
            voice = listener.listen(source)
            rec = listener.recognize_google(voice, language='es-AR')
            rec = rec.lower()
            if name in rec:
                print(rec)
    except:
        pass
    return rec

def run():
    rec = listen()
    if 'reproduce' in rec:
        print('Reproduciendo')
```
